```python
#import
import numpy as np
import scipy.optimize as opt
from scipy.interpolate import interp1d
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import scipy
from scipy import stats
import glob

%matplotlib inline

def chisq(predicted, observed, errors):
    res = observed-predicted
    norm_res = res / errors
    return np.sum(norm_res**2)
```

```python
Vg = np.linspace(0, 2, int(1e6))
ID = (1e-6) / (100e-9) / 2 * 3e-6 * (1e-6 * 100e-9) *  200 * (Vg - 0.8) ** 2
plt.plot(Vg, ID)
plt.xlabel("Voltage (V)", fontsize = 12, fontweight = "bold")
plt.ylabel("Current (A)", fontsize = 12, fontweight = "bold")
```
